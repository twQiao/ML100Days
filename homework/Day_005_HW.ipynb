```python
import requests
import json

#https://www.learncodewithmike.com/2020/09/7-tips-to-avoid-getting-blocked-while-scraping.html

headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9", 
    "Accept-Encoding": "gzip, deflate, br", 
    "Accept-Language": "zh-TW,zh;q=0.9", 
    "Host": "www.dcard.tw",  #目標網站 
    "Sec-Fetch-Dest": "document", 
    "Sec-Fetch-Mode": "navigate", 
    "Sec-Fetch-Site": "none", 
    "Upgrade-Insecure-Requests": "1", 
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36" #使用者代理
}

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true', headers=headers)
r.text
```

```python
#2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

data = json.loads(r.text)
_commentCount=0
_likeCount=0
_totoalCount=0

for d in data:
    _totoalCount = _totoalCount+1
    _likeCount = _likeCount + d['likeCount']
    _commentCount =  _commentCount + d['commentCount']
    print(d['createdAt'] + '>>'+  d['title'] + ' | 留言:' + str(d['commentCount']) + '人 | Like:' + str(d['likeCount']) )
    
print('總筆數:' + str(_totoalCount) + ' | 總留言:' + str(_commentCount) + '人 | 總Like:' + str(_likeCount) + '人 ' )

print('平均留言:' + str( _commentCount/_totoalCount) + ' | 平均Like:' + str( _likeCount/_totoalCount) )
```

```python
#3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false', headers=headers)

data2 = json.loads(r2.text)
_commentCount2=0
_likeCount2=0
_totoalCount2=0

for d in data2:
    _totoalCount2 = _totoalCount2+1
    _likeCount2 = _likeCount2 + d['likeCount']
    _commentCount2 =  _commentCount2 + d['commentCount']
    
print('總筆數:' + str(_totoalCount2) + ' | 總留言:' + str(_commentCount2) + '人 | 總Like:' + str(_likeCount2) )

print('平均留言:' + str( _commentCount2/_totoalCount2) + ' | 平均Like:' + str( _likeCount2/_totoalCount2) )
```
